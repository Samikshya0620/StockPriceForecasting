```python
plt.title('Close History')
plt.plot(data['Date'],data['Close'])
plt.grid(True)
plt.xlabel('Date',fontsize =18)
plt.ylabel('Close Price',fontsize =18)
```

```python
plt.figure(figsize=(8,6))
plt.title('Bollinger Bands')
#plt.plot(data['Date'],data['Close'], label='Close')
plt.plot(data['Date'],data['Bollinger_Middle'], label='Middle')
plt.plot(data['Date'],data['Bollinger_Upper'], label='Upper')
plt.plot(data['Date'],data['Bollinger_Lower'], label='Lower')
plt.legend()
plt.xlabel('Date', fontsize=18)
plt.ylabel('Bollinger Index', fontsize=18)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(30, 30))
sns.heatmap(data.corr(), annot=True, linewidths=0.5)  
plt.title('Correlation Heatmap')

plt.show()
```

```python
plt.figure(figsize=(12, 8)) 
selected_columns = ['Book Value',  'Open',
       'High', 'Low', 'Close','MA_10', 'Bollinger_Middle', 'Bollinger_Upper',
       'Bollinger_Lower']
plt.xticks(rotation=45)
sns.boxplot(data=data[selected_columns], orient="v", palette="Set3") 
```

```python
plt.figure(figsize=(12, 8))  # Set the figure size
selected_columns = ['EPS', 'COF', 'Dividend', 'inflation',
                     'Growth in percent(%)','Percent Change',
                     'MACD', 'Signal_Line', 'ADX']

sns.boxplot(data=data[selected_columns], orient="v", palette="Set3")  # "v" for vertical orientation
plt.ylim(-100,100)

plt.title('Boxplot of Different Columns')
plt.xticks(rotation=45)

# Display the boxplot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))  # Set the figure size
selected_columns = [ 'polarity', 'subjectivity']

sns.boxplot(data=data[selected_columns], orient="v", palette="Set3")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data['Close'], shade=True, color='blue')
plt.title('Density Plot of Close Price')
plt.xlabel('Close Price')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#30-day simple moving average (SMA)
data['SMA_30'] = data['Close'].rolling(window=30).mean()
plt.figure(figsize=(12, 6))
plt.plot(data['Date'],data['Close'], label='Close Price')
plt.plot(data['Date'],data['SMA_30'], label='30-Day SMA', linestyle='--', color='orange')
plt.title('Time Series Data with 30-Day Simple Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
#Simple moving average (SMA)
data['SMA_30'] = data['Close'].rolling(window=30).mean()
data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_20'] = data['Close'].rolling(window=20).mean()
# Plot the original data and the SMA
plt.plot(data['SMA_10'], label='10-Day SMA', color='blue')
plt.plot(data['SMA_20'], label='20-Day SMA',color='orange')
plt.plot(data['SMA_30'], label='30-Day SMA', linestyle='--', color='red')
plt.title('Time Series Data with Simple Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

axes[0, 0].plot(data.index, data['Close'], label='Close Price', color='green')
axes[0, 0].set_xlabel('Date')
axes[0, 0].set_ylabel('Price')
axes[0, 0].legend()

data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data['SMA_30'] = data['Close'].rolling(window=30).mean()


axes[0, 1].plot(data['Date'], data['SMA_10'], label='10-Day SMA', color='blue')
axes[0, 1].set_xlabel('Date')
axes[0, 1].set_ylabel('Price')
axes[0, 1].legend()


axes[1, 0].plot(data['Date'], data['SMA_20'], label='20-Day SMA', color='orange')
axes[1, 0].set_xlabel('Date')
axes[1, 0].set_ylabel('Price')
axes[1, 0].legend()


axes[1, 1].plot(data['Date'], data['SMA_30'], label='30-Day SMA', color='red')
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Price')
axes[1, 1].legend()

# Set a common title for all the subplots
fig.suptitle('Time Series Data with Moving Averages')

# Display the subplots
plt.grid(True)
plt.show()
```
